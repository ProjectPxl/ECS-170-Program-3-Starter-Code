```python
import numpy as np
import matplotlib.pyplot as plt

r = np.load('rewards-itr.npy')
l = np.load('losses-itr.npy')
```

```python
x_val = [x[0] for x in r]
y_val = [x[1] for x in r]
plt.title('Rewards', fontsize="20")
plt.xlabel('frame id', fontsize="20")
plt.ylabel('reward', fontsize="20")
plt.plot(x_val,y_val)
```
